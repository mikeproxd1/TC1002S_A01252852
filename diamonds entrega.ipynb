```python
import pandas as pd
import seaborn as sns
sns.set_style("whitegrid")
sns.set_context("notebook", font_scale=1, rc={"lines.linewidth": 2,'font.family': [u'times']})

import matplotlib.pylab as plt
import numpy as np
```

```python
# Leer los datos a analaizar
diamonds = pd.read_csv("Diamonds are forever.csv")
print('shape: ', diamonds.shape)
diamonds.head()
```

```python
diamonds.columns
```

```python
diamonds.index
```

```python
print ("Min en Carat", diamonds['Carat'].min())
print ("Min en Colour", diamonds['Colour'].min())
print ("Min en Clarity", diamonds['Clarity'].min())
print ("Min en Lab", diamonds['Lab'].min())
print ("Min en Price ($US)", diamonds['Price ($US)'].min())
```

```python
print ("Max en Carat", diamonds['Carat'].max())
print ("Max en Colour", diamonds['Colour'].max())
print ("Max en Clarity", diamonds['Clarity'].max())
print ("Max en Lab", diamonds['Lab'].max())
print ("Max en Price ($US)", diamonds['Price ($US)'].max())
```

```python
print ("Media en Carat", diamonds['Carat'].mean())
print ("Media en Colour", diamonds['Colour'].mean())
print ("Media en Clarity", diamonds['Clarity'].mean())
print ("Media en Lab", diamonds['Lab'].mean())
print ("Media en Price ($US)", diamonds['Price ($US)'].mean())
```

```python
print ("Mediana en Carat", diamonds['Carat'].median())
print ("Mediana en Colour", diamonds['Colour'].median())
print ("Mediana en Clarity", diamonds['Clarity'].median())
print ("Mediana en Lab", diamonds['Lab'].median())
print ("Mediana en Price ($US)", diamonds['Price ($US)'].median())
```

```python
print ("Standar Deviation en Carat", diamonds['Carat'].std())
print ("Standar Deviation en Colour", diamonds['Colour'].std())
print ("Standar Deviation en Clarity", diamonds['Clarity'].std())
print ("Standar Deviation en Lab", diamonds['Lab'].std())
print ("Standar Deviation en Price ($US)", diamonds['Price ($US)'].std())
```
